```python
import numpy as np
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

## clean data 

```python
file_name = "train.csv"
df = pd.read_csv(file_name, sep=",")
df
```

```python
#Target :  
Y = df.y
Y
#Features :  
X = df
X=X.drop(['y'], axis=1)
train_X, valid_X, train_Y, valid_Y = train_test_split( X, Y, test_size=1/7.0, random_state=0)
```

```python
# Get list of categorical variables
s = (train_X.dtypes == 'object')
object_cols = list(s[s].index)
# Make copy to avoid changing original data 
label_X_train = train_X.copy()
label_X_valid = valid_X.copy()

# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for col in object_cols:
    label_X_train[col] = label_encoder.fit_transform(train_X[col])
    label_X_valid[col] = label_encoder.transform(valid_X[col])

print("MAE from Approach 2 (Label Encoding):") 
```

```python
#Missing values  :
# Imputation
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(label_X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(label_X_valid))

# Imputation removed column names; put them back
imputed_X_train.columns = label_X_train.columns
imputed_X_valid.columns = label_X_valid.columns
```

```python
imputed_X_train
```

## PCA analysis

```python
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(imputed_X_train)
# Apply transform to both the training set and the test set.
train_X = scaler.transform(imputed_X_train)
test_X = scaler.transform(imputed_X_valid)
```

```python
# Make an instance of the Model
pca = PCA(.95)
imputed_X_train=pca.fit_transform(imputed_X_train)
```

```python
pca.n_components_ 
```

```python
imputed_X_valid=pca.transform(imputed_X_valid)
imputed_X_valid
```

## Multiple regression analysis 

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(imputed_X_train, train_Y)
print(regr.coef_)
```

```python
r_sq = regr.score(imputed_X_valid, valid_Y)
print('coefficient of determination:', r_sq)
```
